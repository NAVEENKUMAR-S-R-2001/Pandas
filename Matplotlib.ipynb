```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
apple_price=[90,100,100,80,200]
mango_price=[35,120,75,50,140]
year=[2018,2019,2020,2021,2022]
```

```python
plt.plot(year,apple_price,':k',
year,mango_price,'g')
plt.xlabel('Year')
plt.ylabel('Stock price')
plt.axis([2017,2023,20,220])
plt.show()
```
