```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

```python
diamond=pd.read_csv("data_diamonds.csv")
diamond.head()
```

```python
#distribution plot
sns.distplot(diamond['price'])
plt.show()
```

```python
sns.distplot(diamond['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamond['price'],color="red")
plt.show()
```

```python
sns.distplot(diamond['price'],color='green',kde=False,bins=5)
plt.show()
```

```python
sns.distplot(diamond['price'],color="orange",vertical=True)
plt.show()
```
